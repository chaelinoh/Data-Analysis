```python
from minisom import MiniSom
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler


selected_columns = ['인증번호', '인증년도', '설립연도', '이사회_임원등', '이사회_근로자대표', '이사회_수혜자', '이사회_지역사회',
                       '이사회_연간개최회수', '지분율1', '지분보유인원', '취약계층근로자(총인원)', '취약계층평균임금', '취약계층평균근로시간',
                       '비취약계층근로자총인원', '비취약계층평균임금', '비취약계층평균근로시간', '유급근로자총인원', '유급근로자평균임금', '유급근로자평균근로시간',
                       '취약계층_제공인원', '비취약계층_제공인원', '사회서비스_총인원수', '일자리창출_금액', '사회서비스제공_금액', '근로자처우개선_금액',
                       '지역사회재투자_금액', '미처분이익잉여금', '매출액_2020', '매출액_2021', '영업이익_2020', '영업이익_2021',
                       '당기순이익_2020', '당기순이익_2021', '정부지원_2020', '정부지원_2021', '기업후원_2020', '기업후원_2021',
                       '일반지원_2020', '일반지원_2021']

data = df[selected_columns]
data.dropna(inplace=True)

# 데이터 정규화
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

# SOM 모델을 초기화
som_shape = (10, 10)
input_len = len(selected_columns)
som = MiniSom(som_shape[0], som_shape[1], input_len, sigma=0.3, learning_rate=0.5)

# 데이터 적합
som.random_weights_init(data_normalized)
som.train_batch(data_normalized, 1000)  # 반복 횟수 설정 (조정 가능)

# 클러스터 예측
clusters = np.array([som.winner(data_normalized[i]) for i in range(len(data_normalized))])


plt.figure(figsize=(10, 10))
plt.pcolor(som.distance_map().T, cmap='bone_r', alpha=0.6)
plt.colorbar()

for i, j in enumerate(clusters):
    plt.text(j[0] + 0.5, j[1] + 0.5, str(i), ha='center', va='center', fontdict={'size': 8})

plt.xticks(np.arange(0.5, som_shape[0] + 0.5), np.arange(0, som_shape[0]))
plt.yticks(np.arange(0.5, som_shape[1] + 0.5), np.arange(0, som_shape[1]))
plt.title('Self-Organizing Map Clustering')
plt.show()
```

```python
data_selected = df[selected_columns].copy()

# NaN 값 삭제
data_selected.dropna(inplace=True)

# 데이터 정규화
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data_selected)

# SOM 모델 초기화
som_shape = (10, 10)
input_len = len(selected_columns)
som = MiniSom(som_shape[0], som_shape[1], input_len, sigma=0.3, learning_rate=0.5)

# 데이터 적합
som.random_weights_init(data_normalized)
som.train_batch(data_normalized, 1000)

# 각 데이터 포인트가 속한 클러스터를 예측
clusters = np.array([som.winner(data_normalized[i]) for i in range(len(data_normalized))])

# 군집 레이블을 1D로 변환
clusters_1d = clusters[:, 0] * som_shape[1] + clusters[:, 1]

# 군집화 결과를 DataFrame에 추가
data_selected['cluster_label'] = clusters_1d

# 군집화 결과 출력
for cluster_num in range(som_shape[0] * som_shape[1]):
    cluster_data = data_selected[data_selected['cluster_label'] == cluster_num]
    print(f"군집 {cluster_num} 데이터:")
    print(cluster_data)
    print("--------------")

# 각 군집의 통계적 정보를 출력
for cluster_num in range(som_shape[0] * som_shape[1]):
    cluster_data = data_selected[data_selected['cluster_label'] == cluster_num]
    print(f"군집 {cluster_num} 통계:")
    print(cluster_data.describe())
    print("--------------")
```

```python


from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer


selected_features = df[['인증번호', '인증년도', '설립연도', '이사회_임원등', '이사회_근로자대표', '이사회_수혜자', '이사회_지역사회',
                       '이사회_연간개최회수', '지분율1', '지분보유인원', '취약계층근로자(총인원)', '취약계층평균임금', '취약계층평균근로시간',
                       '비취약계층근로자총인원', '비취약계층평균임금', '비취약계층평균근로시간', '유급근로자총인원', '유급근로자평균임금', '유급근로자평균근로시간',
                       '취약계층_제공인원', '비취약계층_제공인원', '사회서비스_총인원수', '일자리창출_금액', '사회서비스제공_금액', '근로자처우개선_금액',
                       '지역사회재투자_금액', '미처분이익잉여금', '매출액_2020', '매출액_2021', '영업이익_2020', '영업이익_2021',
                       '당기순이익_2020', '당기순이익_2021', '정부지원_2020', '정부지원_2021', '기업후원_2020', '기업후원_2021',
                       '일반지원_2020', '일반지원_2021']]


scaler = StandardScaler()
selected_features_scaled = scaler.fit_transform(selected_features)


imputer = SimpleImputer(strategy='mean')
selected_features_imputed = imputer.fit_transform(selected_features_scaled)

# PCA를 적용하여 차원 축소
pca = PCA(n_components=2)
selected_features_pca = pca.fit_transform(selected_features_imputed)

# K-Means 모델을 초기화
k = 4
kmeans_model = KMeans(n_clusters=k, random_state=42)


kmeans_model.fit(selected_features_pca)
clusters = kmeans_model.predict(selected_features_pca)

df['Cluster'] = clusters
print(df)


# 각 군집의 정보 출력
for cluster_id in range(k):
    print(f"\nCluster {cluster_id}:")
    cluster_data = df[df['Cluster'] == cluster_id]
    print(cluster_data.describe())


# 시각화를 위해 주성분으로 변환한 데이터를 산점도로 그리기.
plt.scatter(selected_features_pca[:, 0], selected_features_pca[:, 1], c=df['Cluster'], cmap='rainbow')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering with PCA')
plt.show()
```
